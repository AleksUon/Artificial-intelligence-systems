````python
# –§-—Ü–∏—è –æ—Ü–µ–Ω–∫–∏ –∫–∞—á–µ—Å—Ç–≤–∞ —Ö—Ä–æ–º–æ—Å–æ–º—ã
def qZ(x, y):
    return (x-3*y+1)/(3*(x**2)+3*(y**2)+1)

# –°—É–º–º–∞—Ä–Ω–æ–µ –∫–∞—á–µ—Å—Ç–≤–æ —Ö—Ä–æ–º–æ—Å–æ–º
def qSumZ(Z):
    return sum(Z)

# –û–±–º–µ–Ω —Ö—Ä–æ–º–æ—Å–æ–º–∞–º–∏ –ø–æ —Å—Ö–µ–º–µ
def exchangeScheme(oldX, oldY, sortedId):
    X = [0 for i in range(4)]
    Y = [0 for i in range(4)]

    X[2] = oldX[sortedId[2]]
    X[3] = oldX[sortedId[2]]

    X[0] = oldX[sortedId[0]]

    X[1] = oldX[sortedId[1]]

    Y[0] = oldY[sortedId[2]]
    Y[1] = oldY[sortedId[2]]

    Y[2] = oldY[sortedId[0]]

    Y[3] = oldY[sortedId[1]]

    return X, Y

# –§-—Ü–∏—è —Å–æ—Ä—Ç–∏—Ä–æ–≤–∫–∏ –º–∞—Å—Å–∏–≤–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –ø–æ—Ç–æ–º–∫–æ–≤ –∏ –≤—ã–¥–µ–ª–µ–Ω–∏–µ –∏–Ω–¥–µ–∫—Å–æ–≤
def sorting(Z):
    sortedId = sorted(range(len(Z)), key = lambda k: Z[k])
    return sortedId

# –§-—Ü–∏—è —à–∞–≥–∞ —ç–≤–æ–ª—é—Ü–∏–∏
def evoStep(X, Y, Z):
    _, minId = min((value, id) for (id, value) in enumerate(Z))
    X = X[:]
    Y = Y[:]
    Z = Z[:]

    X.pop(minId)
    Y.pop(minId)
    Z.pop(minId)

    return X, Y, Z

# —à–∞–≥–∏ —ç–≤–æ–ª—é—Ü–∏–∏ (–∫–æ–Ω–µ—á–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è), 4 —à–∞–≥–∞
def evoSteps(X, Y, stepsNum = 4):
    results = []

    for i in range(4):
        arrZ = [qZ(x, Y[i]) for i, x in enumerate(X)]
        X, Y, Z = evoStep(X, Y, arrZ)

        X, Y = exchangeScheme(X, Y, sorting(Z))

        results.append([X, Y, qSumZ(arrZ), arrZ])
    return X, Y, results

# –û–±—ä—è–≤–ª–µ–Ω–∏–µ –º–∞—Å—Å–∏–≤–æ–≤ —Ö—Ä–æ–º–æ—Å–æ–º
X = [-2, -1, 0, 1]
Y = [-2, -1, 0, 1]

# –†–µ–∞–ª–∏–∑–∞—Ü–∏—è –∞–ª–≥–æ—Ä–∏—Ç–º–∞
results = evoSteps(X, Y)

# –í—ã–≤–æ–¥ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤
for i in range(len(results[2])):
    print(f'max_{i+1}_step: {results[2][i][2]}')

qualityArrZ = []
for i in range(len(results[2])):
    qualityArrZ += results[2][i][3]
print(f'max Z:     {max(qualityArrZ)}')
````

````python
import networkx as nx
from math import e
import matplotlib.pyplot as plt

# –û–ø–∏—Å–∞–Ω–∏–µ –º–∞—Å—Å–∏–≤–∞ –¥–ª–∏–Ω—ã —Ä–µ–±–µ—Ä, –ø–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç–∏ –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –≤–µ—Ä—à–∏–Ω, –∏—Ö –∑–∞–º–µ–Ω—ã, –∑–Ω–∞—á–µ–Ω–∏–µ –†, –∞ —Ç–∞–∫–∂–µ –Ω–∞—á–∞–ª—å–Ω—É—é —Ç–µ–º–ø—É
distances = [
    (1, 2, 20),
    (1, 3, 40),
    (1, 4, 42),
    (1, 5, 33),
    (1, 6, 21),
    (2, 3, 26),
    (2, 4, 38),
    (2, 5, 42),
    (2, 6, 17),
    (3, 4, 22),
    (3, 5, 43),
    (3, 6, 21),
    (4, 5, 27),
    (4, 6, 22),
    (5, 6, 26)] # –î–ª–∏–Ω—ã —Ä–µ–±–µ—Ä

V = [1, 4, 5, 2, 6, 3, 1] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞
Z = [(3, 4), (4, 6), (5, 2), (6, 2)] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω—ã –≤–µ—Ä—à–∏–Ω
P = [49, 54, 43, 54] # –°–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å—á—ë—Ç–∞
T = 100 # –ù–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–∞

# –§-—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
def probability(delta, T):
    return 100 * e ** (-delta / T)

# –§-—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø—ã
def reductTemp(prevT):
    nextT = 0.5 * prevT
    return nextT

graph = nx.Graph() # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞
graph.add_weighted_edges_from(distances) # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä–µ–±–µ—Ä

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)

#–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞
def edgeLength(i, j, distances, roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1)
                    for item in distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –º–∞—Ä—à—Ä—É—Ç–∞
def routeLength(V, distances):
    edges = []
    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))
    return sum(edges)

# –û–¥–Ω–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏
def routeOneReplacement(arrV, Z, replacementByName = True):
    decrement = 1 if replacementByName else 0
    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]
    return arrV

# –ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V

# –í—ã–±–æ—Ä –Ω—É–¥–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances) # –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏
    arrSum = [sumLength] # –ú–∞—Å—Å–∏–≤ —Å—É–º–º –¥–ª–∏–Ω

    # –¶–ò–∫–ª –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i]) # –ù–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances) # –î–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
        arrSum.append(newS)
        deltaS = newS - sumLength # –†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–æ–≤

        # –í —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –Ω–∏–º–∏ –±–æ–ª—å—à–µ 0, —Ç–æ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
        if deltaS > 0:
            p = probability(deltaS, T)

            # –ï—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ p, —Ç–æ –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è
            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
        T = reductTemp(T) # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç–µ–º–ø—ã

    return V, arrSum

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –º–∞—Ä—à—Ä—É—Ç—É
def drawRouteGraph(distances, bestRoute):
    newDistances = []
    # –ü—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏–µ –ø–æ –≤–µ–∫—Ç–æ—Ä—É
    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)

    graph = nx.Graph() # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞

    graph.add_weighted_edges_from(newDistances) # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä–µ–±–µ—Ä

    # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
    nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–Ω–∏–µ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞ –∏ –µ–≥–æ –¥–ª–∏–Ω—ã
bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

# –°–æ–∑–¥–∞–Ω–∏–µ –¥–≤—É—Ö –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤
fig, (ax1) = plt.subplots(1, figsize=(8, 6))

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –ª—É—á—à–∏–º –º–∞—Ä—à—Ä—É—Ç–æ–º
drawRouteGraph(distances, bestRoute)

plt.tight_layout()
plt.show()
````

````python
import networkx as nx
from math import e
import matplotlib.pyplot as plt

distances = [
    (1, 2, 25),
    (1, 3, 41),
    (1, 4, 38),
    (1, 5, 27),
    (1, 6, 20),
    (2, 3, 21),
    (2, 4, 34),
    (2, 5, 39),
    (2, 6, 17),
    (3, 4, 24),
    (3, 5, 40),
    (3, 6, 22),
    (4, 5, 21),
    (4, 6, 21),
    (5, 6, 22)] # –î–ª–∏–Ω—ã —Ä–µ–±–µ—Ä

V = [1, 3, 5, 4, 6, 2, 1] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –ø—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏—è –º–∞—Ä—à—Ä—É—Ç–∞
Z = [(3, 4), (4, 6), (5, 6), (2, 4)] # –ü–æ—Å–ª–µ–¥–æ–≤–∞—Ç–µ–ª—å–Ω–æ—Å—Ç—å –∑–∞–º–µ–Ω—ã –≤–µ—Ä—à–∏–Ω
P = [41, 60, 85, 60] # –°–ª—É—á–∞–π–Ω—ã–µ —á–∏—Å–ª–∞, –≤—ã–ø–∞–≤—à–∏–µ –≤ –ø—Ä–æ—Ü–µ—Å—Å–µ —Å—á—ë—Ç–∞
T = 100 # –ù–∞—á–∞–ª—å–Ω–∞—è —Ç–µ–º–ø–∞

# –§-—Ü–∏—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–∏
def probability(delta, T):
    return 100 * e ** (-delta / T)

# –§-—Ü–∏—è –∏–∑–º–µ–Ω–µ–Ω–∏—è —Ç–µ–º–ø—ã
def reductTemp(prevT):
    nextT = 0.5 * prevT
    return nextT

graph = nx.Graph() # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞
graph.add_weighted_edges_from(distances) # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä–µ–±–µ—Ä

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)

#–í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã —Ä–µ–±—Ä–∞
def edgeLength(i, j, distances, roundTrip = True):
    if roundTrip:
        return max([(item[2] if (item[0] == i and item[1] == j) or (item[1] == i and item[0] == j) else -1)
                    for item in distances])
    else:
        return max([(item[2] if (item[0] == i and item[1] == j) else -1) for item in distances])

# –í—ã—á–∏—Å–ª–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –º–∞—Ä—à—Ä—É—Ç–∞
def routeLength(V, distances):
    edges = []
    for i in range(len(V) - 1):
        edges.append(edgeLength(V[i], V[i + 1], distances))
    return sum(edges)

# –û–¥–Ω–∞ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏
def routeOneReplacement(arrV, Z, replacementByName = True):
    decrement = 1 if replacementByName else 0
    arrV[Z[0] - decrement], arrV[Z[1] - decrement] = arrV[Z[1] - decrement], arrV[Z[0] - decrement]
    return arrV

# –ü–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∞ –≤ –ø—É—Ç–∏
def routeReplacement(V, Z):
    for z in Z:
        V = routeOneReplacement(V, z)
    return V

# –í—ã–±–æ—Ä –Ω—É–¥–Ω–æ–≥–æ –ø—É—Ç–∏ –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
def chooseRoute(distances, V, Z, T, P):
    sumLength = routeLength(V, distances) # –ù–∞—Ö–æ–∂–¥–µ–Ω–∏–µ –¥–ª–∏–Ω—ã –ø—É—Ç–∏
    arrSum = [sumLength] # –ú–∞—Å—Å–∏–≤ —Å—É–º–º –¥–ª–∏–Ω

    # –¶–ò–∫–ª –º–µ—Ç–æ–¥–æ–º –æ—Ç–∂–∏–≥–∞
    for i in range(len(Z)):
        newV = routeOneReplacement(V[:], Z[i]) # –ù–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –ø–æ—Å–ª–µ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏
        newS = routeLength(newV, distances) # –î–ª–∏–Ω–∞ –Ω–æ–≤–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞
        arrSum.append(newS)
        deltaS = newS - sumLength # –†–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –¥–ª–∏–Ω–æ–π –Ω–æ–≤–æ–≥–æ –∏ —Å—Ç–∞—Ä–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–æ–≤

        # –í —Å–ª—É—á–∞–µ, –µ—Å–ª–∏ —Ä–∞–∑–Ω–∏—Ü–∞ –º–µ–∂–¥—É –Ω–∏–º–∏ –±–æ–ª—å—à–µ 0, —Ç–æ –≤—ã—á–∏—Å–ª—è–µ—Ç—Å—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å
        if deltaS > 0:
            p = probability(deltaS, T)

            # –ï—Å–ª–∏ –∑–∞–¥–∞–Ω–Ω–∞—è –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç—å –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –∏–Ω—Ç–µ—Ä–≤–∞–ª –æ—Ç 0 –¥–æ p, —Ç–æ –Ω–æ–≤—ã–π –º–∞—Ä—à—Ä—É—Ç –≤—ã–±–∏—Ä–∞–µ—Ç—Å—è
            if p > P[i]:
                V = newV
                sumLength = newS
        else:
            V = newV
            sumLength = newS
        T = reductTemp(T) # –í—ã—á–∏—Å–ª–µ–Ω–∏–µ —Ç–µ–º–ø—ã

    return V, arrSum

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ –ø–æ –∑–∞–¥–∞–Ω–Ω–æ–º—É –º–∞—Ä—à—Ä—É—Ç—É
def drawRouteGraph(distances, bestRoute):
    newDistances = []
    # –ü—Ä–æ—Ö–æ–∂–¥–µ–Ω–∏–µ –ø–æ –≤–µ–∫—Ç–æ—Ä—É
    for i in range(len(bestRoute) - 1):
        for distance in distances:
            if distance[0] == bestRoute[i] and distance[1] == bestRoute[i + 1] or distance[1] == bestRoute[i] and distance[0] == bestRoute[i + 1]:
                newDistances.append(distance)

    graph = nx.Graph() # –°–æ–∑–¥–∞–Ω–∏–µ –ø—É—Å—Ç–æ–≥–æ –≥—Ä–∞—Ñ–∞

    graph.add_weighted_edges_from(newDistances) # –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Ä–µ–±–µ—Ä

    # –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –∑–∞–¥–∞–Ω–Ω—ã–º–∏ –≤–µ—Ä—à–∏–Ω–∞–º–∏
    nx.draw_kamada_kawai(graph, node_color = '#fb7258', node_size = 2000, with_labels = True)

# –†–∞—Å—Å—á–∏—Ç—ã–≤–∞–Ω–∏–µ –ª—É—á—à–µ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞ –∏ –µ–≥–æ –¥–ª–∏–Ω—ã
bestRoute, arrLength = chooseRoute(distances, V, Z, T, P)

print(f'–õ—É—á—à–∏–π –≤—ã–±—Ä–∞–Ω–Ω—ã–π –º–∞—Ä—à—Ä—É—Ç: {bestRoute}')
print(f'–î–ª–∏–Ω–∞ –ª—É—á—à–µ–≥–æ –≤—ã–±—Ä–∞–Ω–Ω–æ–≥–æ –º–∞—Ä—à—Ä—É—Ç–∞: {routeLength(bestRoute, distances)}')
print(f'–î–ª–∏–Ω—ã –≤—Å–µ—Ö —Ä–∞—Å—Å–º–æ—Ç—Ä–µ–Ω–Ω—ã—Ö –º–∞—Ä—à—Ä—É—Ç–æ–≤: {arrLength}')

# –°–æ–∑–¥–∞–Ω–∏–µ –¥–≤—É—Ö –æ—Ç–¥–µ–ª—å–Ω—ã—Ö –≥—Ä–∞—Ñ–∏–∫–æ–≤
fig, (ax1) = plt.subplots(1, figsize=(8, 6))

# –û—Ç—Ä–∏—Å–æ–≤–∫–∞ –≥—Ä–∞—Ñ–∞ —Å –ª—É—á—à–∏–º –º–∞—Ä—à—Ä—É—Ç–æ–º
drawRouteGraph(distances, bestRoute)

plt.tight_layout()
plt.show()
````
